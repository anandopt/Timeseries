```python
ndf = df.drop(['Date_Time'], axis=1)
ndf.index=df.Date_Time
ndf.shape[1]
```

```python
nanindex=df.isnull(any(1))
```

```python
ndf.index
```

```python
# The -200 represents missing values. This value is replaced with its value previous day

cols=ndf.columns
for j in cols:
    for i in range(len(ndf)):
        ndf[ndf[j]==-200] = ndf[j][i-1]
```

```python
# getting dataframe with no NaT in index
ndf=ndf[ndf.index.notnull()]
ndf.columns
```

```python
ndf.drop('Unnamed: 15', axis=1, inplace=True)
ndf.drop('Unnamed: 16', axis=1, inplace=True)
```

```python
from statsmodels.tsa.vector_ar.vecm import coint_johansen
johan_test_temp = ndf.drop(['CO(GT)'], axis=1)
print(coint_johansen(johan_test_temp, -1,1).eig)
```

```python
# Creating train and test sets
train = ndf[:int(0.8*(len(df)))]
test=ndf[int(0.8*len(df)):]
```

```python
from statsmodels.tsa.vector_ar.var_model import VAR
model = VAR(endog=train)
model_fit=model.fit()
```

```python
prediction=model_fit.forecast(model_fit.y, steps=len(test))
prediction
```

```python
pred = pd.DataFrame(index=range(0,len(prediction)),columns=[cols])
for j in range(0,13):
    for i in range(0, len(prediction)):
       pred.iloc[i][j] = prediction[i][j]

    
#check rmse
from sklearn.metrics import mean_squared_error
for i in cols:
    print('rmse value for', i, 'is : ', sqrt(mean_squared_error(pred[i], test[i])))
```
